```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Initialize StandardScaler
standard_scaler = StandardScaler()

# Apply scaling
standard_scaled = standard_scaler.fit_transform(data)

# Convert to DataFrame for easier interpretation
standard_scaled_df = pd.DataFrame(standard_scaled, columns=data.columns)

print("\nStandard Scaled Data:")
print(standard_scaled_df)
```

```python
# Initialize MinMaxScaler
min_max_scaler = MinMaxScaler()

# Apply scaling
min_max_scaled = min_max_scaler.fit_transform(data)

# Convert to DataFrame for easier interpretation
min_max_scaled_df = pd.DataFrame(min_max_scaled, columns=data.columns)

print("\nMin-Max Scaled Data:")
print(min_max_scaled_df)
```

# Key Differences
  
1)Standard Scaling:
* Centers the data around 0 with a standard deviation of 1.
* Useful for algorithms sensitive to the magnitude of data, like SVM or Logistic Regression.
  
2)Min-Max Scaling:
* Scales all values to a fixed range, typically [0, 1].
* Useful for algorithms like Neural Networks where scaled inputs help in faster convergence.
