```python
import pandas as pd
```

```python
train = pd.read_csv("Titanic_train.csv")
train
```

```python
test = pd.read_csv("Titanic_test.csv")
test
```

```python
print(train.info())
```

```python
print(test.info())
```

```python
train = train.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
train
```

```python
test = test.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
test
```

```python
train.isna().sum()
```

```python
test.isna().sum()
```

```python
train=train.dropna()
train.isna().sum()
```

```python
test=test.dropna()
test.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['Sex']=le.fit_transform(train['Sex'])
train.head()
```

```python
test['Sex']=le.fit_transform(test['Sex'])
test.head()
```

```python
train['Embarked']=le.fit_transform(train['Embarked'])
train.head()
```

```python
test['Embarked']=le.fit_transform(test['Embarked'])
test.head()
```

```python
x_tr = x = train.iloc[:,1:]
x_tr
```

```python
y_tr = y = train.iloc[:,0]
```

```python
y_tr
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=32)
```

```python
x_train,x_test,y_train,y_test
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Train XGBoost
xgb = XGBClassifier(n_estimators=50, learning_rate=0.1, max_depth=1, random_state=42, use_label_encoder=False, eval_metric='mlogloss')
xgb.fit(x_train, y_train)

# Predictions
y_pred = xgb.predict(x_test)
y_pred
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred))
```

```python
y_train_pred = xgb.predict(x_train)
y_train_pred
print("XGBoost Accuracy:", accuracy_score(y_train, y_train_pred))
```

```python
from lightgbm import LGBMClassifier
import lightgbm as lgb


# Train LightGBM
lgb = LGBMClassifier(n_estimators=50, learning_rate=0.1, max_depth=1, random_state=42)
lgb.fit(x_train, y_train)

# Predictions
y_pred = lgb.predict(x_test)
y_pred
print("LightGBM Accuracy:", accuracy_score(y_test, y_pred))
```

```python
y_train_pred_l = lgb.predict(x_train)
y_train_pred_l
print("LightGBM Accuracy:", accuracy_score(y_train, y_train_pred_l))
```
