```python
import numpy as np
from sklearn.model_selection import KFold 
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
# Initialize the Random Forest Classifier
rf_clf = RandomForestClassifier(n_estimators=100, random_state=8)
# Fit the model on the entire dataset
rf_clf.fit(x, y)
# Predict the target values
y_pred = rf_clf.predict(x)
```

```python
y_pred
```

```python
k = 10  # Replace with your desired number of folds

# Initialize k-Fold
kf = KFold(n_splits=k, shuffle=True, random_state=8)

# Perform cross-validation
cv_scores = cross_val_score(rf_clf, x, y, cv=kf, scoring='accuracy')

# Print cross-validation results
print(f'Cross-validation scores: {cv_scores}')
print(f'Mean accuracy: {np.mean(cv_scores):.2f}')
print(f'Standard deviation: {np.std(cv_scores):.2f}')
```

```python
print('Accuracy:', accuracy_score(y, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y, y_pred))
print('Classification Report:\n', classification_report(y, y_pred))
```
